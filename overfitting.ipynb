```python
import matplotlib.pyplot as plt
import numpy as np

# Close all existing figures to clear previous plots
plt.close("all")

# Define the overfitting example function
def overfit_example(show_plot=True):
    # Generate sample data
    x = np.linspace(-3, 3, 100)
    y = np.sin(x) + np.random.normal(0, 0.2, size=x.shape)

    # Overfitting model (polynomial regression)
    coefficients = np.polyfit(x, y, deg=10)  # High-degree polynomial
    y_fit = np.polyval(coefficients, x)

    # Create the plot
    fig, ax = plt.subplots(figsize=(8, 6))
    ax.scatter(x, y, label="Data", color="blue", alpha=0.6)
    ax.plot(x, y_fit, label="Overfitted Model", color="red")
    ax.legend()
    ax.set_title("Overfitting Example")
    ax.set_xlabel("X")
    ax.set_ylabel("Y")

    # Show or return the figure based on the argument
    if show_plot:
        plt.show()
    else:
        return fig

# Execute the function without displaying the plot
ofit = overfit_example(False)

# Display the figure (alternative to `display(output)`)
if ofit:
    ofit.show()
```
