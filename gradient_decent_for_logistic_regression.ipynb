```python
import copy
import math
import numpy as np
import matplotlib.pyplot as plt

# Enable interactive plotting in Jupyter Notebook
%matplotlib widget

# Use a built-in Matplotlib style
plt.style.use('ggplot')  # Alternative: 'seaborn-darkgrid', 'classic', etc.

# Sigmoid function
def sigmoid(z):
    """Compute the sigmoid of z."""
    return 1 / (1 + np.exp(-z))

# Compute logistic regression cost
def compute_cost_logistic(X, y, w, b):
    """Compute cost for logistic regression."""
    m = X.shape[0]
    cost = 0
    for i in range(m):
        z = np.dot(X[i], w) + b
        f_wb = sigmoid(z)
        cost += -y[i] * np.log(f_wb) - (1 - y[i]) * np.log(1 - f_wb)
    return cost / m

# Compute logistic regression gradient
def compute_gradient_logistic(X, y, w, b):
    """Compute gradients for logistic regression."""
    m, n = X.shape
    dj_dw = np.zeros(n)
    dj_db = 0

    for i in range(m):
        z = np.dot(X[i], w) + b
        f_wb = sigmoid(z)
        error = f_wb - y[i]
        dj_dw += error * X[i]
        dj_db += error

    dj_dw /= m
    dj_db /= m

    return dj_dw, dj_db

# Plot data points
def plot_data(X, y):
    """Plot data points with labels."""
    plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], marker='o', label="Positive", c='b')
    plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], marker='x', label="Negative", c='r')
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.legend()

# Plot decision boundary
def plot_decision_boundary(X, y, w, b):
    """Plot decision boundary for logistic regression."""
    plot_data(X, y)
    
    x_values = np.array([min(X[:, 0]) - 1, max(X[:, 0]) + 1])
    y_values = -(w[0] * x_values + b) / w[1]
    
    plt.plot(x_values, y_values, 'g', label="Decision Boundary")
    plt.legend()
    plt.show()
```

```python
X_train=np.array([[0.5,1.5],[1,1],[1.5,0.5],[3,0.5],[2,2],[1,2.5]])
y_train=np.array([0,0,0,1,1,1])
```

```python
fig,ax=plt.subplots(1,1,figsize=(4,4))
plot_data(X_train,y_train)

ax.axis([0,4,0,3.5])
ax.set_ylabel('$x_1$',fontsize=12)
ax.set_xlabel('$x_0$',fontsize=12)
```
