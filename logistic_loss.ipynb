```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train=np.array([0.,1,2,3,4,5],dtype=np.longdouble)
y_train=np.array([0,0,0,1,1,1],dtype=np.longdouble)
```

```python
x_lower = x_train[y_train == 0]  # Points where y = 0
y_lower = y_train[y_train == 0]

x_upper = x_train[y_train == 1]  # Points where y = 1
y_upper = y_train[y_train == 1]

# Plot
plt.scatter(x_lower, y_lower, color='blue', label='y = 0')
plt.scatter(x_upper, y_upper, color='red', label='y = 1')

plt.xlabel("x_train")
plt.ylabel("y_train")
plt.title("Scatter Plot of Training Data")
plt.yticks([0, 1])  # Ensure y-axis only has 0 and 1
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

```python
# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Squared error cost function for logistic regression
def cost_function(w, b, x_train, y_train):
    m = len(x_train)
    total_cost = 0
    for i in range(m):
        f_wb = sigmoid(w * x_train[i] + b)  # Compute prediction
        total_cost += (f_wb - y_train[i]) ** 2  # Squared error loss
    return total_cost / (2 * m)  # Compute mean squared error

# Function to plot cost for different values of w and b
def plot_cost_function(x_train, y_train):
    w_values = np.linspace(-5, 5, 50)  # Range for w
    b_values = np.linspace(-5, 5, 50)  # Range for b

    W, B = np.meshgrid(w_values, b_values)  # Create a grid
    Cost = np.zeros_like(W)

    # Compute cost for each (w, b) pair
    for i in range(W.shape[0]):
        for j in range(W.shape[1]):
            Cost[i, j] = cost_function(W[i, j], B[i, j], x_train, y_train)

    # Plot 3D cost surface
    fig = plt.figure(figsize=(10, 6))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(W, B, Cost, cmap='viridis')

    ax.set_xlabel("w")
    ax.set_ylabel("b")
    ax.set_zlabel("Cost J(w, b)")
    ax.set_title("Squared Error Cost Function for Logistic Regression")

    plt.show()

# Example data
x_train = np.array([0., 1, 2, 3, 4, 5], dtype=np.longdouble)
y_train = np.array([0, 0, 0, 1, 1, 1], dtype=np.longdouble)

# Plot cost function
plt.close('all')
plot_cost_function(x_train, y_train)
```

```python
# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Squared error loss function
def squared_error(y_true, y_pred):
    return (y_true - y_pred) ** 2

# Function to plot two logistic loss curves
def plt_two_logistic_loss_curves():
    plt.figure(figsize=(8, 5))  # Set figure size
    
    # Generate logistic predictions (for visualization purposes)
    z = np.linspace(-10, 10, 100)  # Range for logistic function
    y_pred = sigmoid(z)  # Predicted probabilities
    
    # Compute squared error for both cases y = 0 and y = 1
    y_true_0 = np.zeros_like(y_pred)  # True label = 0
    y_true_1 = np.ones_like(y_pred)   # True label = 1
    
    loss_0 = squared_error(y_true_0, y_pred)  # Loss for y = 0
    loss_1 = squared_error(y_true_1, y_pred)  # Loss for y = 1

    # Plot squared error loss
    plt.plot(y_pred, loss_0, label="Squared Error (y = 0)", color='blue')
    plt.plot(y_pred, loss_1, label="Squared Error (y = 1)", color='red')

    plt.xlabel("Predicted Probability")
    plt.ylabel("Squared Error Loss")
    plt.title("Squared Error Loss Curves for Logistic Regression")
    plt.legend()
    plt.grid(True)

# Clear previous plots
plt.close('all')

# Call the function to plot
plt_two_logistic_loss_curves()
plt.show()
```

```python
# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Logistic Regression Cost Function using Squared Error
def logistic_cost(w, b, x_train, y_train):
    m = len(x_train)  # Number of samples
    total_cost = 0
    for i in range(m):
        f_wb = sigmoid(w * x_train[i] + b)  # Compute prediction
        total_cost += (f_wb - y_train[i]) ** 2  # Squared error loss
    return total_cost / (2 * m)  # Compute mean squared error

# Function to compute and plot cost function
def plt_logistic_cost(x_train, y_train):
    w_values = np.linspace(-5, 5, 100)  # Range for w
    b_values = np.linspace(-5, 5, 100)  # Range for b

    W, B = np.meshgrid(w_values, b_values)  # Create a grid
    Cost = np.zeros_like(W)

    # Compute cost for each (w, b) pair
    for i in range(W.shape[0]):
        for j in range(W.shape[1]):
            Cost[i, j] = logistic_cost(W[i, j], B[i, j], x_train, y_train)

    # Store the minimum cost
    min_cost = np.min(Cost)

    # Plot the cost function surface
    fig = plt.figure(figsize=(10, 6))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(W, B, Cost, cmap='viridis')

    ax.set_xlabel("w")
    ax.set_ylabel("b")
    ax.set_zlabel("Cost J(w, b)")
    ax.set_title("Squared Error Cost Function for Logistic Regression")

    plt.show()

    return min_cost  # Return the minimum cost for reference

# Example dataset
x_train = np.array([0., 1, 2, 3, 4, 5], dtype=np.longdouble)
y_train = np.array([0, 0, 0, 1, 1, 1], dtype=np.longdouble)

# Clear previous plots
plt.close('all')

# Compute and plot cost function
cst = plt_logistic_cost(x_train, y_train)
print("Minimum Cost Value:", cst)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Logistic loss function
def logistic_loss(w, b, x_train, y_train):
    m = len(x_train)  # Number of training examples
    total_loss = 0

    for i in range(m):
        f_wb = sigmoid(w * x_train[i] + b)
        loss = - (y_train[i] * np.log(f_wb) + (1 - y_train[i]) * np.log(1 - f_wb))
        total_loss += loss

    return total_loss / m  # Mean loss

# Function to plot loss function
def plt_two_logistic_loss_curves(x_train, y_train):
    w_values = np.linspace(-5, 5, 100)  # Range of w values
    b_values = np.linspace(-5, 5, 100)  # Range of b values
    W, B = np.meshgrid(w_values, b_values)  # Create a grid
    Loss = np.zeros_like(W)

    # Compute loss for each (w, b) pair
    for i in range(W.shape[0]):
        for j in range(W.shape[1]):
            Loss[i, j] = logistic_loss(W[i, j], B[i, j], x_train, y_train)

    # 3D plot of logistic loss
    fig = plt.figure(figsize=(12, 6))
    ax = fig.add_subplot(121, projection='3d')
    ax.plot_surface(W, B, Loss, cmap='coolwarm')

    ax.set_xlabel("w")
    ax.set_ylabel("b")
    ax.set_zlabel("Log Loss J(w, b)")
    ax.set_title("Logistic Loss Surface")

    # 2D heatmap of loss function
    ax2 = fig.add_subplot(122)
    contour = ax2.contourf(W, B, Loss, cmap='coolwarm')
    fig.colorbar(contour, ax=ax2)
    
    ax2.set_xlabel("w")
    ax2.set_ylabel("b")
    ax2.set_title("Logistic Loss Contour Map")

    plt.show()

# Example dataset
x_train = np.array([0., 1, 2, 3, 4, 5], dtype=np.longdouble)
y_train = np.array([0, 0, 0, 1, 1, 1], dtype=np.longdouble)

# Clear previous plots
plt.close('all')

# Compute and plot loss function
plt_two_logistic_loss_curves(x_train, y_train)
```
