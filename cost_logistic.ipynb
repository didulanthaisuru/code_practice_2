```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X_train = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])  #(m,n)
y_train = np.array([0, 0, 0, 1, 1, 1])                                           #(m,)
```

```python
def plot_data(X, y, ax):
    ax.scatter(X[y == 0][:, 0], X[y == 0][:, 1], color='blue', label='Class 0', marker='o')
    ax.scatter(X[y == 1][:, 0], X[y == 1][:, 1], color='red', label='Class 1', marker='x')

# Create plot
fig, ax = plt.subplots(1, 1, figsize=(4, 4))

# Plot the data
plot_data(X_train, y_train, ax)

# Set both axes to be from 0-4
ax.axis([0, 4, 0, 3.5])

# Labels and title
ax.set_ylabel('$x_1$', fontsize=12)
ax.set_xlabel('$x_0$', fontsize=12)

# Show plot
plt.show()
```

```python
def compute_cost_logistic(X,y,w,b):
    m=X.shape[0]
    cost=0.0
    for i in range(m):
        z_i=np.dot(X[i],w)+b
        f_wb_i=sigmoid(z_i)
        cost+= -y[i]*np.log(f_wb_i)-(1-y[i])*np.log(1-f_wb_i)
    cost=cost/m
    return cost
```

```python
w_tmp=np.array([1,1])
b_tmp=-3
print(compute_cost_logistic(X_train,y_train,w_tmp,b_tmp))
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the color dictionary
dlc = {
    "dlblue": "blue",
    "dlmagenta": "magenta"
}

# Data
X_train = np.array([[0.5, 1.5], [1, 1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y_train = np.array([0, 0, 0, 1, 1, 1])

# Define the plotting function
def plot_data(X, y, ax):
    ax.scatter(X[y == 0][:, 0], X[y == 0][:, 1], color='blue', label='Class 0', marker='o')
    ax.scatter(X[y == 1][:, 0], X[y == 1][:, 1], color='red', label='Class 1', marker='x')

# Choose values between 0 and 6 for x0
x0 = np.arange(0, 6)

# Plot the two decision boundaries
x1 = 3 - x0  # Decision boundary for b = -3
x1_other = 4 - x0  # Decision boundary for b = -4

# Create plot
fig, ax = plt.subplots(1, 1, figsize=(4, 4))

# Plot the decision boundaries
ax.plot(x0, x1, c=dlc["dlblue"], label="$b$=-3")
ax.plot(x0, x1_other, c=dlc["dlmagenta"], label="$b$=-4")
ax.axis([0, 4, 0, 4])

# Plot the original data
plot_data(X_train, y_train, ax)

# Set axis labels and title
ax.set_ylabel('$x_1$', fontsize=12)
ax.set_xlabel('$x_0$', fontsize=12)

# Show the legend and title
plt.legend(loc="upper right")
plt.title("Decision Boundary")
plt.show()
```
